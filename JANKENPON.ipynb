```python
#Script para joguinho rápido de JAN KEN PON (esse é o nome correto do jogo), em três rodadas.
#Este jogo é uma adaptação do código original que me foi apresentado pelo professor e mestre Gleisson Bispo.
#O jogo original fazia uma disputa de JANKENPON e apresentava o vencedor entre a máquina e o usuário.
#Eu acrescentei um pouco de inteligência e aleatoriedade e criei uma "melhor de 3".

from random import choice
from time import sleep
jogadas = 0
vitorias_user = 0
vitorias_pc = 0

print('VAMOS VER QUEM GANHA NO JAN KEN PON, EM 3 TENTATIVAS!!!\n')

while jogadas != 3:
    escolha_pc = choice(['PEDRA', 'PAPEL', 'TESOURA'])
    escolha_user = input('ESCOLHA PEDRA, PAPEL OU TESOURA:').upper().strip()
    resultado_pc = choice(['HAHAHA... EU GANHEI! :D', 'EU GANHEI! EU GANHEI! :P', 'HIHIHI... VOCÊ PERDEU! ;)'])
    resultado_user = choice(['AHH... VOCÊ GANHOU... :(', 'NÃO VALE... VOCÊ GANHOU... >:(', 'ASSIM NÃO DÁ... VOCÊ GANHOU... :,('])
    print('JAN...')
    sleep(1)
    print('KEN...')
    sleep(1)
    print('PON!\n')
    if escolha_pc == escolha_user:
        print(f'EU ESCOLHI {escolha_pc}!')
        print(f'VOCÊ ESCOLHEU {escolha_user}!')
        print('XIII... DEU EMPATE...')
        jogadas += 1
        
    elif escolha_pc == 'PEDRA' and escolha_user == 'PAPEL':
        print(f'EU ESCOLHI {escolha_pc}!')
        print(f'VOCÊ ESCOLHEU {escolha_user}!')
        print(resultado_user)
        jogadas += 1
        vitorias_user += 1
        
    elif escolha_pc == 'PEDRA' and escolha_user == 'TESOURA':
        print(f'EU ESCOLHI {escolha_pc}!')
        print(f'VOCÊ ESCOLHEU {escolha_user}!')
        print(resultado_pc)
        jogadas += 1
        vitorias_pc += 1
        
    elif escolha_pc == 'PAPEL' and escolha_user == 'TESOURA':
        print(f'EU ESCOLHI {escolha_pc}!')
        print(f'VOCÊ ESCOLHEU {escolha_user}!')
        print(resultado_user)
        jogadas += 1
        vitorias_user += 1
        
    elif escolha_pc == 'PAPEL' and escolha_user == 'PEDRA':
        print(f'EU ESCOLHI {escolha_pc}!')
        print(f'VOCÊ ESCOLHEU {escolha_user}!')
        print(resultado_pc)
        jogadas += 1
        vitorias_pc += 1
        
    elif escolha_pc == 'TESOURA' and escolha_user == 'PEDRA':
        print(f'EU ESCOLHI {escolha_pc}!')
        print(f'VOCÊ ESCOLHEU {escolha_user}!')
        print(resultado_user)
        jogadas += 1
        vitorias_user += 1
     
    elif escolha_pc == 'TESOURA' and escolha_user == 'PAPEL':
        print(f'EU ESCOLHI {escolha_pc}!')
        print(f'VOCÊ ESCOLHEU {escolha_user}!')
        print(resultado_pc)
        jogadas += 1
        vitorias_pc += 1
        
    if jogadas < 3:
        print('VAMOS TENTAR DE NOVO...\n')
        
if vitorias_pc == vitorias_user:
    print('NINGUÉM GANHOU...')
    
elif vitorias_pc > vitorias_user:
    print('UHUUU... EU SOU O GRANDE VENCEDOR!')
    
elif vitorias_pc < vitorias_user:
    print('VOCÊ É O GRANDE CAMPEÃO!!!')
```
